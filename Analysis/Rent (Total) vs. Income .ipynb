```python
#Data cleanup here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import csv
import json
import requests
import seaborn as sns
import time
```

```python
# Pull csv using URL of dataset uploaded to repo
url = 'https://raw.githubusercontent.com/calderon0423/Commitment-Issues/master/Dataset/'

all_homes = pd.read_csv(f'{url}State_MedianRentalPrice_AllHomes.csv')
del all_homes['Unnamed: 0']
all_homes.head()
```

```python
#find the sum of rents for each state in 2015
all_homes_2015 = all_homes[['RegionName', 'SizeRank', '2015-01', '2015-02', '2015-03', '2015-04', '2015-05', '2015-06',
                           '2015-07', '2015-08', '2015-09', '2015-10', '2015-11', '2015-12',
                           ]]
all_homes_2015 = all_homes_2015.dropna()
sum_2015 = all_homes_2015[['2015-01', '2015-02', '2015-03', '2015-04', '2015-05', '2015-06',
                           '2015-07', '2015-08', '2015-09', '2015-10', '2015-11', '2015-12',
                           ]]

sum_2015["2015 Median"] = sum_2015.median(axis=1)
sum_2015['2015 Total'] = round(sum_2015["2015 Median"] * 12, 2)
sum_2015["RegionName"] = all_homes_2015["RegionName"]
sum_2015 = sum_2015[["RegionName", "2015 Total"]]
sum_2015.head()
sum_2015
```

```python
all_homes_2016 = all_homes[['RegionName', '2016-01', '2016-02', '2016-03', '2016-04', '2016-05', '2016-06',
                           '2016-07', '2016-08', '2016-09', '2016-10', '2016-11', '2016-12',
                           ]]
all_homes_2016 = all_homes_2016.dropna()
sum_2016 = all_homes_2016[['2016-01', '2016-02', '2016-03', '2016-04', '2016-05', '2016-06',
                           '2016-07', '2016-08', '2016-09', '2016-10', '2016-11', '2016-12'
                           ]]

sum_2016['2016 Median'] = sum_2016.median(axis=1)
sum_2016['2016 Total'] = round(sum_2016['2016 Median'] * 12, 2)
sum_2016["RegionName"] = all_homes_2016["RegionName"]
sum_2016 = sum_2016[["RegionName", "2016 Total"]]
sum_2016.head()
```

```python
all_homes_2017 = all_homes[['RegionName', '2017-01', '2017-02', '2017-03', '2017-04', '2017-05', '2017-06',
                           '2017-07', '2017-08', '2017-09', '2017-10', '2017-11', '2017-12',
                           ]]
all_homes_2017 = all_homes_2017.dropna()
sum_2017 = all_homes_2017[['2017-01', '2017-02', '2017-03', '2017-04', '2017-05', '2017-06',
                           '2017-07', '2017-08', '2017-09', '2017-10', '2017-11', '2017-12'
                           ]]

sum_2017['2017 Median'] = sum_2017.median(axis=1)
sum_2017['2017 Total'] = round(sum_2017['2017 Median'] * 12, 2)
sum_2017["RegionName"] = all_homes_2017["RegionName"]
sum_2017 = sum_2017[["RegionName", "2017 Total"]]
sum_2017.head()
```

```python
#Merge dataframes
new = pd.merge(sum_2015, sum_2016, on="RegionName")
new = pd.merge(new, sum_2017, on="RegionName")
sorted_2015_high = new.sort_values("2015 Total", ascending=False)
sorted_2015_high.head(3)
```

```python
sorted_2015_low = new.sort_values("2015 Total")
sorted_2015_low.head(3)
```

```python
sorted_2016_high = new.sort_values("2016 Total", ascending=False)
sorted_2016_high.head(3)
```

```python
sorted_2016_low = new.sort_values("2016 Total")
sorted_2016_low.head(3)
```

```python
sorted_2017_high = new.sort_values("2017 Total", ascending=False)
sorted_2017_high.head(3)
```

```python
sorted_2017_low = new.sort_values("2017 Total")
sorted_2017_low.head(3)
```

```python
new["'15 - '16 Change"] = new["2016 Total"] - new["2015 Total"]
new["'16 - '17 Change"] = new["2017 Total"] - new["2016 Total"]
renamed = new.rename(columns={"RegionName":"State",
                                "2015 Total":"2015 Yearly Rent",
                                "2016 Total":"2016 Yearly Rent",
                                 "2017 Total":"2017 Yearly Rent",
                                 "'15 - '16 Change":"'15-'16 Rent Diff.",
                                 "'16 - '17 Change":"'16-'17 Rent Diff."})
renamed
```

```python
income_df = pd.read_csv("../household_median_income_2017.csv")
income_df = income_df.dropna(how='any')

#excluting Nebraska & United States rows to match the Zillow df. Puerto Rico was not in csv file. 
income_df = income_df.loc[income_df['State']!= 'Nebraska']
income_df = income_df.loc[income_df['State']!= 'United States'].reset_index(drop=True)
income_df = income_df[['State','2017', '2016', '2015']].sort_values('2017',ascending = False)

#removing commas & converting to int data type 
income_df["2015 Income"] = income_df["2015"].str.replace(",","").astype(int)
income_df["2016 Income"] = income_df["2016"].str.replace(",","").astype(int)
income_df["2017 Income"] = income_df["2017"].str.replace(",","").astype(int)

#replace so matches rent
income_df = income_df.drop(columns={"2017", "2016", "2015"})
income_df["State"] = income_df["State"].replace({"D.C.":"District of Columbia"})

#find difference
income_df["'15-'16 Income Diff."] = income_df['2016 Income'] - income_df['2015 Income']
income_df["'16-'17 Income Diff."] = income_df['2017 Income'] - income_df['2016 Income']

income_df = income_df.reset_index(drop=True)
income_df
```

```python
merged = pd.merge(renamed, income_df, on="State")
merged = merged.sort_values("State").reset_index(drop=True)
merged
```

```python
merged.to_csv("../output_data/rental(total)_income_merged.csv", index=False, header=True)
```
