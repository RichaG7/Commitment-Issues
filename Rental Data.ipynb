```python
#Data cleanup here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import csv
import json
import requests
import seaborn as sns
```

```python
data = './Dataset/State_MedianRentalPrice_AllHomes.csv'
rent_price_df = pd.read_csv(data)
del rent_price_df['Unnamed: 0']
#rent_price_df.drop(rent_price_df.columns[[0, 4, 2]], axis = 1, inplace = True) 
rent_price_df2 = rent_price_df.drop(rent_price_df.loc[:, '2010-01':'2015-12'].columns, axis = 1) 
rent_price_2016_2017 = rent_price_df2.drop(rent_price_df2.loc[:, '2018-01':'2019-12'].columns, axis = 1) 
states_info = rent_price_2016_2017[rent_price_2016_2017.columns[0:2]]
#rent_price_2016 = rent_price_20162017.drop('2017-01','2017-02','2017-03', '2017-04', '2017-05', '2017-06','2017-07', '2017-08', '2017-09', '2017-10', '2017-11', '2017-12')                    
rent_price_2016 = rent_price_2016_2017.drop(list(rent_price_2016_2017)[14:26], axis=1)
rent_price_2017 = rent_price_2016_2017.drop(list(rent_price_2016_2017)[2:14], axis=1)

#rent_price_avg 2016 
rent_price_2016['2016_avg'] = round(rent_price_2016.mean(axis = 1, skipna = True), 2)
#rent_price_avg 2017
rent_price_2017['2017_avg'] = round(rent_price_2017.mean(axis = 1, skipna = True), 2)

#sorted for LOWESR rent state (2016 & 2017)
highest_rent2017 = rent_price_2017.sort_values('2017_avg', ascending=False)
highest_rent2016 = rent_price_2016.sort_values('2016_avg', ascending=False)

#sorted for LOWEST rent state (2016 & 2017)
lowest_rent2017 = rent_price_2017.sort_values('2017_avg', ascending=True)
lowest_rent2016 = rent_price_2016.sort_values('2016_avg', ascending=True)

```

```python
#highest_rent2016.set_index('RegionName', inplace=True)

#2=--2016combining states with avg(deleted month break down)
Region2016 = rent_price_2016.iloc[:, 0:2].reset_index()
Average2016 = rent_price_2016.iloc[:, -1:].reset_index()

#--2017 combining states with avg(deleted month break down)
Region2017 = rent_price_2017.iloc[:, 0:2].reset_index()
Average2017 = rent_price_2017.iloc[:, -1:].reset_index()

#result = pd.merge(Region,Average[['use_id', 'platform', 'device']], on='use_id', how='left')
merge2016 = pd.merge(Region2016,Average2016, on='index', how='outer')
merge2017 = pd.merge(Region2017,Average2017, on='index', how='outer')
average_2017 = Average2017.iloc[:, -1].reset_index()

merge2017 = pd.DataFrame(average_2017)
final_merge = pd.merge(merge2016,merge2017, on='index', how='right')
del final_merge['index']
final_merge

by_2016 = final_merge.sort_values(by=['2016_avg'], ascending=False)
by_2017 = final_merge.sort_values(by=['2017_avg'], ascending=False)

```
