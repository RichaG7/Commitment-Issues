```python
#Data cleanup here

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import csv
import json
import requests
import seaborn as sns
```

```python
studio = pd.read_csv('MedianRent_Studio.csv')
del studio['Unnamed: 0']
studio = pd.melt(studio, id_vars=['RegionName', 'State', 'Metro', 'CountyName', 'SizeRank'])
studio.columns = ['RegionName', 'State', 'Metro', 'CountyName', 'SizeRank', 'Time',
       'Value']
studio['Type'] = 'studio'
studio = studio.loc[studio['Time'].isin(['2012-01', '2012-02', '2012-03', '2012-04', 
                                         '2012-05', '2012-06','2012-07', '2012-08', 
                                         '2012-09', '2012-10', '2012-11', '2012-12',
                                         '2018-01', '2018-02', '2018-03', '2018-04',
                                         '2018-05', '2018-06','2018-07', '2018-08', 
                                         '2018-09', '2018-10', '2018-11', '2018-12'])]


one_bedroom = pd.read_csv('MedianRent_1Bedroom.csv')
del one_bedroom['Unnamed: 0']
one_bedroom = pd.melt(one_bedroom, id_vars=['RegionName', 'State', 'Metro', 'CountyName', 'SizeRank'])
one_bedroom.columns = ['RegionName', 'State', 'Metro', 'CountyName', 'SizeRank', 'Time',
       'Value']
one_bedroom['Type'] = 'one bedroom'
one_bedroom = one_bedroom.loc[one_bedroom['Time'].isin(['2012-01', '2012-02', '2012-03', '2012-04', 
                                                   '2012-05', '2012-06','2012-07', '2012-08', 
                                                   '2012-09', '2012-10', '2012-11', '2012-12',
                                                   '2018-01', '2018-02', '2018-03', '2018-04', 
                                                   '2018-05', '2018-06','2018-07', '2018-08', 
                                                   '2018-09', '2018-10', '2018-11', '2018-12'])]


two_bedroom = pd.read_csv('MedianRent_2Bedroom.csv')
del two_bedroom['Unnamed: 0']
two_bedroom = pd.melt(two_bedroom, id_vars=['RegionName', 'State', 'Metro', 'CountyName', 'SizeRank'])
two_bedroom.columns = ['RegionName', 'State', 'Metro', 'CountyName', 'SizeRank', 'Time',
       'Value']
two_bedroom['Type'] = 'two bedroom'
two_bedroom = two_bedroom.loc[two_bedroom['Time'].isin(['2012-01', '2012-02', '2012-03', '2012-04', '2012-05', '2012-06',
                                                  '2012-07', '2012-08', '2012-09', '2012-10', '2012-11', '2012-12',
                                                  '2018-01', '2018-02', '2018-03', '2018-04', '2018-05', '2018-06',
                                                  '2018-07', '2018-08', '2018-09', '2018-10', '2018-11', '2018-12'])]

three_bedroom = pd.read_csv('MedianRent_3Bedroom.csv')
del three_bedroom['Unnamed: 0']
three_bedroom = pd.melt(three_bedroom, id_vars=['RegionName', 'State', 'Metro', 'CountyName', 'SizeRank'])
three_bedroom.columns = ['RegionName', 'State', 'Metro', 'CountyName', 'SizeRank', 'Time',
       'Value']
three_bedroom['Type'] = 'three bedroom'
three_bedroom = three_bedroom.loc[three_bedroom['Time'].isin(['2012-01', '2012-02', '2012-03', '2012-04', '2012-05', '2012-06',
                                                      '2012-07', '2012-08', '2012-09', '2012-10', '2012-11', '2012-12',
                                                      '2018-01', '2018-02', '2018-03', '2018-04', '2018-05', '2018-06',
                                                      '2018-07', '2018-08', '2018-09', '2018-10', '2018-11', '2018-12'])]

four_bedroom = pd.read_csv('MedianRent_4Bedroom.csv')
del four_bedroom['Unnamed: 0']
four_bedroom = pd.melt(four_bedroom, id_vars=['RegionName', 'State', 'Metro', 'CountyName', 'SizeRank'])
four_bedroom.columns = ['RegionName', 'State', 'Metro', 'CountyName', 'SizeRank', 'Time',
       'Value']
four_bedroom['Type'] = 'four bedroom'
four_bedroom = four_bedroom.loc[four_bedroom['Time'].isin(['2012-01', '2012-02', '2012-03', '2012-04', '2012-05', '2012-06',
                                                    '2012-07', '2012-08', '2012-09', '2012-10', '2012-11', '2012-12',
                                                    '2018-01', '2018-02', '2018-03', '2018-04', '2018-05', '2018-06',
                                                    '2018-07', '2018-08', '2018-09', '2018-10', '2018-11', '2018-12'])]

five_bedroom = pd.read_csv('MedianRent_5BedroomOrMore.csv')
del five_bedroom['Unnamed: 0']
five_bedroom = pd.melt(five_bedroom, id_vars=['RegionName', 'State', 'Metro', 'CountyName', 'SizeRank'])
five_bedroom.columns = ['RegionName', 'State', 'Metro', 'CountyName', 'SizeRank', 'Time',
       'Value']
five_bedroom['Type'] = 'five bedroom'
five_bedroom = five_bedroom.loc[five_bedroom['Time'].isin(['2012-01', '2012-02', '2012-03', '2012-04', '2012-05', '2012-06',
                                                    '2012-07', '2012-08', '2012-09', '2012-10', '2012-11', '2012-12',
                                                    '2018-01', '2018-02', '2018-03', '2018-04', '2018-05', '2018-06',
                                                    '2018-07', '2018-08', '2018-09', '2018-10', '2018-11', '2018-12'])]

sfr = pd.read_csv('MedianRent_Sfr.csv')
del sfr['Unnamed: 0']
sfr = pd.melt(sfr, id_vars=['RegionName', 'State', 'Metro', 'CountyName', 'SizeRank'])
sfr.columns = ['RegionName', 'State', 'Metro', 'CountyName', 'SizeRank', 'Time',
       'Value']
sfr['Type'] = 'sfr'
sfr = sfr.loc[sfr['Time'].isin(['2012-01', '2012-02', '2012-03', '2012-04', '2012-05', '2012-06',
                                  '2012-07', '2012-08', '2012-09', '2012-10', '2012-11', '2012-12',
                                  '2018-01', '2018-02', '2018-03', '2018-04', '2018-05', '2018-06',
                                  '2018-07', '2018-08', '2018-09', '2018-10', '2018-11', '2018-12'])]

all_homes = pd.read_csv('MedianRent_AllHomes.csv')
del all_homes['Unnamed: 0']
all_homes = pd.melt(all_homes, id_vars=['RegionName', 'State', 'Metro', 'CountyName', 'SizeRank'])
all_homes.columns = ['RegionName', 'State', 'Metro', 'CountyName', 'SizeRank', 'Time',
       'Value']
all_homes['Type'] = 'all homes'
all_homes = all_homes.loc[all_homes['Time'].isin(['2012-01', '2012-02', '2012-03', '2012-04', '2012-05', '2012-06',
                                              '2012-07', '2012-08', '2012-09', '2012-10', '2012-11', '2012-12',
                                              '2018-01', '2018-02', '2018-03', '2018-04', '2018-05', '2018-06',
                                              '2018-07', '2018-08', '2018-09', '2018-10', '2018-11', '2018-12'])]




```

```python
merged_df = studio.append(one_bedroom)
merged_df = merged_df.append(two_bedroom)
merged_df = merged_df.append(three_bedroom)
merged_df = merged_df.append(four_bedroom)
merged_df = merged_df.append(five_bedroom)
merged_df = merged_df.append(sfr)
merged_df = merged_df.append(all_homes)
merged_df = merged_df.dropna()
merged_df['Time'] = pd.to_datetime(merged_df['Time']) 

merged_df_2012 = merged_df.loc[(merged_df['Time'] >= '2012-01-01') & (merged_df['Time'] <= '2012-12-01')]

merged_df_2018 = merged_df.loc[(merged_df['Time'] >= '2018-01-01') & (merged_df['Time'] <= '2018-12-01')]


```

```python
grouped_df_2012 = merged_df_2012.groupby(['RegionName', 'Time']).agg(
    median_rentalprice = pd.NamedAgg(column='Value', aggfunc='median')).reset_index()

grouped_df_2012_top = grouped_df_2012.loc[(grouped_df_2012['median_rentalprice']>= 3000)]

sns.lmplot(x="Time", y="median_rentalprice", data=grouped_df_2012_top, fit_reg=False, hue='RegionName', legend=False)


# Move the legend to an empty part of the plot
plt.legend(loc='upper center', ncol=5)
plt.ylim(3000, 7000)
plt.show()


```

```python
grouped_df_2018 = merged_df_2018.groupby(['RegionName', 'Time']).agg(
    median_rentalprice = pd.NamedAgg(column='Value', aggfunc='median')).reset_index()

grouped_df_2018_top = grouped_df_2018.loc[(grouped_df_2018['median_rentalprice']>= 10000)]

sns.lmplot(x="Time", y="median_rentalprice", data=grouped_df_2018_top, fit_reg=False, hue='RegionName', legend=False)


# Move the legend to an empty part of the plot
plt.legend(loc='upper center', ncol=5)
plt.ylim(10000, 65000)
plt.show()
```

```python
grouped_df_2012_bottom = grouped_df_2012.loc[(grouped_df_2012['median_rentalprice']<= 700)]

sns.lmplot(x="Time", y="median_rentalprice", data=grouped_df_2012_bottom, fit_reg=False, hue='RegionName', legend=False)


# Move the legend to an empty part of the plot
plt.legend(loc='upper center', ncol=5)
plt.ylim(500, 1000)
plt.show()

```

```python
grouped_df_2018_bottom = grouped_df_2018.loc[(grouped_df_2018['median_rentalprice']<= 700)]

sns.lmplot(x="Time", y="median_rentalprice", data=grouped_df_2018_bottom, fit_reg=False, hue='RegionName', legend=False)

# Move the legend to an empty part of the plot
plt.legend(loc='upper center', ncol=5)
plt.ylim(250, 1000)
plt.show()
```
